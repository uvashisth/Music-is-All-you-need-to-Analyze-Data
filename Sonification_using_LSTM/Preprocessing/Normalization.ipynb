```python
df = pd.DataFrame(data)
df.head()
```

```python
values = df.iloc[:,1:2]
```

```python
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(50,90))
x_scaled = min_max_scaler.fit_transform(values)
new_df = pd.DataFrame(x_scaled)
```

```python
print(new_df)
```

```python
new_df = np.ceil(new_df)
new_df
```

```python
convert_list = new_df.values.tolist()
convert_list
```

```python
 new_list = []
for i in convert_list:
    for j in i:
        new_list.append(j)
print( new_list)
```
